Code given below uses sklearn libraries to load the dataset,slpit the dataset into train and test sets then this dataset will be trained through knn classifier.
accuracy_score will evaluate the performance on the testing data.At the end matplotlib.pyplot will visualize the dataset.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

Iris flower dataset is loaded and is stored in object named iris. X is also an object that has all features form iris dataset and Y has the target vector that is representing the type of flower.

```python
iris = load_iris()
X, y = iris.data, iris.target
```

It is splitting the dataset into training and testing sets. X_train is a training feature matrix that is 80% of the data, X_test is the testing feature matrix that is 20% of the data.
Y_train is the training target vector and Y_test is the testing target vector.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

train_acc , test_accu are two empty lists that will store the training and testing accuracy. Ks is an object which defines the range of k-values from 1 to 15 for knn algorithm.

```python
train_acc, test_acc = [], []
ks = range(1, 16)
```

These iterations will execute until the value of k meets the range.
Trains KNN classifier with k neighbors.
Then fits the model to training data and appends the accuracy of training to the train_acc object after that it appends the accuracy of testing to the test_acc. This loop optimizes the KNN model with different K-values and tracks its performance on the training and testing data.

```python
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    train_acc.append(knn.score(X_train, y_train))
    test_acc.append(knn.score(X_test, y_test))
```

It creates different plots like
1.training accuracy vs k-values
2.testing accuracy vs k-values
with labels axes, title, grids etc to visualize the effect of K on KNN Accuracy

```python
plt.figure(figsize=(8, 5))
plt.plot(ks, train_acc, marker='o', label="Training Accuracy")
plt.plot(ks, test_acc, marker='s', label="Testing Accuracy")
plt.xlabel("Value of K")
plt.ylabel("Accuracy")
plt.title("Effect of K on KNN Accuracy (Iris Dataset)")
plt.legend()
plt.grid(True)
plt.show()
```
